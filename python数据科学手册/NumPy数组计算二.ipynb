## 广播的规则

- 1. 如果两个数组的维度数不相同，那么小维度数组的形状将会在最左边补1.

- 2. 如果两个数组的形状在任何一个维度上都不匹配，那么数组的形状会沿着维度为1的维度扩展以匹配另外一个数组的形状。

- 3. 如果两个数组的形状在任何一个维度上都不匹配并且没有任何一个维度等于1.那么会引发异常。

### 示例一 

```python
M = np.ones((2, 3))
a = np.arange(3)
```

分析步骤：
1. 查看数组形状
M.shape = (2, 3)
a.shape = (3,)

2. 根据规则1，数组a的维度数更小，在其左边补1，补1后的形状为：
a.shape = (1, 3)

3. 根据规则2，数组的形状沿着维度数为1的维度扩展，匹配另外一个数组的形状，扩展后的形状为
a.shape = (2, 3)

4. 此时，进行运算
M + a

```python
M+a
```

### 示例二

```python
a = np.arange(3).reshape((3,1))
b = np.arange(3)
```

分析步骤：

1. 写出两个数组的形状
a.shape = (3, 1)
b.shape = (3,)

2. 根据规则1，数组a的维度数更小，在其左边补1，补1后的形状为： b.shape = (1, 3), a.shape=(3, 1)

3. 根据规则2，数组的形状沿着维度为1的维度扩展，以匹配另一个数组，扩展的数组形状为： a.shape = (3, 3), b.shape = (3, 3)

```python
a + b
```

### 示例三

```python
M = np.ones((3, 2))
a = np.arange(3)
```

分析步骤：

1. 分析数组形状，a.shape = (3,), M.shape = (3, 2)

2. 根据规则1，a的形状必须用1进行补全，M.shape = (3, 2), a.shape = (1, 3);

3. 根据规则2，扩展a的维度为1的维度匹配M的维度, a.shape = (3, 3), M.shape = (3, 2)

**最终形状还是不匹配，因此这两个数组是不兼容的**

```python
M + a
```
